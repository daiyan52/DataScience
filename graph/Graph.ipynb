```python
x = np.arange(1,10)
y = x ** 3
plt.plot(x,y,'pink')
plt.show()
```

```python
x = np.arange(1,10,0.3)
y = x ** 3

y1 = x**2
# plt.plot(x,y,color='red',marker='o')
plt.plot(x,y,color='black',linewidth='1.5',label='x^3')
plt.plot(x,y1,color='red',linewidth='1',label='x^2')
plt.axis([0,10,0,100])
plt.legend()
plt.ylabel('y')
plt.xlabel('x')
plt.text(8,80,'abc',fontsize=20,color='blue')
plt.grid()
plt.title('Example_1')
plt.show()
```
