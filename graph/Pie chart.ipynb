```python
import matplotlib.pyplot as plt
```

```python
size = [1,2,4,5]
labels = ['A','B','C','D']
colors =['pink','yellow','red','blue']
explode = [0.1,0,0,0]
plt.title('Pie Chart')
plt.axis('equal')
plt.pie(size,colors=colors,labels=labels,autopct='%0.2f%%',explode=explode,startangle=180,counterclock=False)
plt.show()
```
