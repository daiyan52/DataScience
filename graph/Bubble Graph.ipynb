```python
import matplotlib.pyplot as plt
```

```python
year = [2010,2012,2014,2016,2018,2020,2022]
salary = [15,18,22,35,45,60,100]
population = [50,90,110,160,190,200,230]
plt.axis([2010,2026,20,250])
plt.title('Bubble Chart')
plt.scatter(year,salary,s=population,color='red',alpha=0.5)
plt.show()
```

```python
import numpy as np
c = np.arange(len(year))
c[2]=c[3]
c
```

```python
plt.scatter(year,salary,s=population,c=c,alpha=0.5)
```

```python
plt.text(year[0],salary[0],50)
plt.scatter(year,salary,s=population,c=c,alpha=0.5,edgecolor='red')
```
