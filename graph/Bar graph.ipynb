### Bar graph

```python
import matplotlib.pyplot as plt
```

```python
student =['IT','CSE','ME','ECE','EE','EIE','CE']
number =[125,300,60,100,45,34,110]
plt.bar(student,number)
plt.show()
```

```python
student =['IT-2024','CSE-2024','ME-2024','ECE-2024','EE-2024','EIE-2024','CE-2024']
number =[125,300,60,100,45,34,110]
plt.xticks(rotation=40,color='red')
plt.bar(student,number,color='black',edgecolor='red',linewidth=1,alpha=0.5)
plt.show()
```
