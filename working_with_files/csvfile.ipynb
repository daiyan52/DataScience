```python
country_killed ={}
for  row in file_list:
    key = row['Country']
    value = row['Killed']
    
    if value!='':
        value = int(float(value))
    else:
        value =0
    if key in country_killed:
        country_killed[key]+=value
    else:
        country_killed[key]=value
country_killed
```

```python
import csv
with open('/Users/daiyanalam/Documents/MachineLearnig/Dataset/year2017.csv') as file_obj:
    file_data = csv.DictReader(file_obj,skipinitialspace=True)
    list_file = list(file_data)
```

```python
type(list_file)
```

```python
list_file
```

```python
month_killed = {}
for row in list_file:
    key = row['Month']
    value = row['Killed']
    
    if value!='':
        value = int(float(value))
    else:
        value =0
    if key in month_killed:
        month_killed[key] = month_killed[key] + value
    else:
        month_killed[key] = value
```
